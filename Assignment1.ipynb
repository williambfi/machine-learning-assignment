## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

Jenis kriminalitas tertinggi dan perkembangannya dalam category yang major di daerah Soutwark

```python
# Write your function below
borough_filter = ['Southwark']

df_drugs_southwark = df[(df['borough'].isin(borough_filter))]

df_area_drugs = df_drugs_southwark[['year','major_category','value']]

df_area_drugs = df_area_drugs.pivot_table(values='value', index = 'year', \
                     columns= 'major_category', aggfunc= sum)
                     
# Graded-Funtion Begin (~1 Lines)

# Graded-Funtion End
df_area_drugs.plot(kind='area', 
             stacked=False,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('Tingkat Krimininalitas yang Major di daerah Southwark')
plt.ylabel('value')
plt.xlabel('year')

plt.show()
```

**Insight:**

jika dilihat dari area plot diatas dapat terlihat bahwa burglary paling banyak terjadi di daerah Soutwark dan kondisi saat ini meningkat.

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**
apa frekuensi distribusi criminalitas dari daerah Southwark,Barnet,Bexley  pada dari tahun 2008 ke 2016?
 

```python

borough_filter = ['Southwark','Barnet','Bexley']

df_histogram = df[(df['borough'].isin(borough_filter))]

df_histogram = df_histogram[['year','borough','value']]

df_histogram = df_histogram.pivot_table(values='value', index = 'year', \
                     columns= 'borough', aggfunc= sum)

df_histogram.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Jumlah Kriminalitas dari daerah Southwark,Barnet dan Bexley dari 2008 - 2016')
plt.ylabel('Jumlah Tahun')
plt.xlabel('Jumlah Kriminalitas')

plt.show()

```

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**
Komparasi kriminalitas category 'Drugs' dari tahun 2008 ke 2016 di daerah Southwark

```python
# Write your function below
borough_filter = ['Southwark']
major_category_filter = ['Drugs']

df_bar = df[(df['borough'].isin(borough_filter)) & (df['major_category'].isin(major_category_filter))]
df_bar = df_bar[['year','value']]

df_bar.set_index('year',inplace=True)

df_sum = df_bar.groupby(['year']).agg('sum')


# Graded-Funtion Begin (~1 Lines)

# Graded-Funtion End
df_sum.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of Kriminalitas') # add y-label to the plot
plt.title('Kriminalitas Category Drugs di daerah Soutwark dari tahun 2008 ke 2016') # add title to the plot

plt.show()
```

**Insight:**

terlihat kalau ada penurunan kriminalitas drugs  dalam 8 tahun terakhir

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

Berapa proporsi kriminalitas kategori major didaerah Soutwark

```python
borough_filter = ['Southwark']

df_pie = df[(df['borough'].isin(borough_filter))]

df_pie = df_pie[['year','major_category','value']]

df_pie = df_pie.pivot_table(values='value', index = 'year', \
                     columns= 'major_category', aggfunc= sum)
df_pie = df_pie.transpose()
df_pie['Total'] = df_pie.sum(axis=1)
df_pie.head()
```

```python
# Write your function below
df_pie['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow      
                            )

plt.title('proposi kriminalitas kategori major dari tahun 2008 ke 2016 di daerah Soutwark')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

**Insight:**

terlihat theft dan handling paling banyak terjadi di daerah southwark dalam 8 tahun terakhir

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**
melihat apakah ada  korelasi antara daerah 'Southwark' dan'Tower Hamlets' untuk kriminalitas Drugs

```python
# # Write your function below
major_category_filter = ['Drugs','Robbery']

# df_scatter = df[(df['major_category'].isin(major_category_filter))]

# df_scatter = df.loc[(df['major_category'].isin(major_category_filter))]

df_scatter = df
df_scatter.loc[df['major_category'] .isin(major_category_filter)] 

df_scatter['date'] = pd.to_datetime(df_scatter[['year', 'month']].assign(DAY=1))
df_scatter = df_scatter[['date','value']]

df_scatter = df_scatter.groupby('date').sum()

df_scatter.reset_index(inplace = True)


# # Graded-Funtion Begin (~1 Lines)
df_scatter.plot(kind='scatter', x='date', y='value', figsize=(10, 6), color='darkblue')

plt.title('korealasi drugs dan roberry pada semua daerah london dari tahun 2008 ke 2016')
plt.xlabel('Year')
plt.ylabel('Number of Criminal')

plt.show()

```

`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
# instantiate a word cloud object
your_wordcloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

df_word =  df[(df['year'] == 2016)]

# generate the word cloud
text = " ".join(i for i in df_word.major_category)
your_wordcloud.generate(text)
```

## Folium

Folium is a powerful Python library that helps you create several types of Leaflet maps. The fact that the Folium results are interactive makes this library very useful for dashboard building.

From the official Folium documentation page:

> Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library. Manipulate your data in Python, then visualize it in on a Leaflet map via Folium.

> Folium makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

> The library has a number of built-in tilesets from OpenStreetMap, Mapbox, and Stamen, and supports custom tilesets with Mapbox or Cloudmade API keys. Folium supports both GeoJSON and TopoJSON overlays, as well as the binding of data to those overlays to create choropleth maps with color-brewer color schemes.

```python
#!conda install -c conda-forge folium=0.5.0 --yes
!pip install folium
import folium

print('Folium installed and imported!')
```

```python
# define the world map
world_map = folium.Map()
```

```python

latitude = 51.509865
longitude = -0.118092

london_boroughs = pd.DataFrame([['Barking & Dagenham', '51.558777647704765', '0.13459701347460404'], 
                ['Barnet', '51.61031041355852', '-0.20535824781288015']], columns=['borough', 'x', 'y'])

london_boroughs = london_boroughs.iloc[0:,:]


london_map = folium.Map(location=[latitude,longitude], zoom_start=8)

criminals = folium.map.FeatureGroup()

for lat, lng, in zip(london_boroughs.x, london_boroughs.y):
    criminals.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=15, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

london_map.add_child(criminals)
london_map
```
